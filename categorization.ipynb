```python
import pandas as pd
```

```python
"""
video ID	an 11-digit string, which is unique
uploader	a string of the video uploader's username
age	an integer number of days between the date when the video was uploaded and Feb.15, 2007 (YouTube's establishment)
category	a string of the video category chosen by the uploader
length	an integer number of the video length
views	an integer number of the views
rate	a float number of the video rate
ratings	an integer number of the ratings
comments	an integer number of the comments
related IDs	up to 20 strings of the related video IDs"""
```

```python
root_path = "C:\\Users\\Ritik's PC\\Desktop\\College 4th Year\\Spring 2022\\CS 436\\youtube-network-aggregation\\YouTubeData\\080327\\"
column_names = ["video_id", "uploader", "age", 'category', "length", "views", "rate", "ratings", "comments",
            "r1", "r2", "r3", "r4", "r5", "r6", "r7", "r8", "r9", "r10", "r11", "r12", "r13", "r14", "r15", "r16", "r17", "r18", "r19", "r20"  
           ]
df0 = pd.read_csv(root_path+"0.txt", sep="\t", names=column_names, header=None)
df1 = pd.read_csv(root_path+"1.txt", sep="\t", names=column_names, header=None)
df2 = pd.read_csv(root_path+"2.txt", sep="\t", names=column_names, header=None)
df3 = pd.read_csv(root_path+"3.txt", sep="\t", names=column_names, header=None)
```

```python
df =pd.concat([df0, df1, df2, df3])
```

```python
df.groupby('category')['video_id'].count().plot(kind="bar")
```

```python
# bar chart by category
# find top ten users wtih maximum uploads
# top ten videos by viewes /
# top ten videos by views/length/ratings/comments in every category
```

```python
df.groupby('category')['video_id'].count().plot(kind="bar")
```

```python
# top 10 users with maximum number of uploads
counts_df = df.groupby('uploader').size().reset_index(name="counts").sort_values(by=["counts"], ascending=False)[:10]
counts_df.plot(x="uploader", y="counts", kind="barh")

# uploader_counts.sort_values(by=["counts"], ascending = False)[:10].plot(kind="bar")
```

```python
buckets = [0,100,1000,10000, 100000, 1000000]
labels = ['<100', '<1000', '<10000', "<100000", "<1000000"]
series = pd.cut(df['views'], buckets, labels=labels)
views_df = series.to_frame()
views_df.groupby('views').size().plot(kind="barh")
```
